# Contagem de modificações de campos por valor final de algum campo específico

## Importações

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline

plt.style.use('ggplot')
pd.set_option('display.float_format', lambda x: '%.2f' % x)


df_bugs_changes = pd.read_json('data\\processed_bugs_counting_changes_final.json')
df_bugs_resolution = pd.read_json('data\\bugs_id_resolution_final.json')
df_bugs_resolution.rename(columns={'resolution': 'final_resolution'}, inplace=True)

df_bugs = df_bugs_changes.merge(df_bugs_resolution, how='inner', on='bug_id')
df_bugs = df_bugs.drop(columns=['bug_id'])
df_bugs['total_changes'] = df_bugs.sum(axis=1, numeric_only=True)
df_bugs = df_bugs[['total_changes', 'final_resolution']]
```

## Agrupamento de contagem* de modificações por resolução final
\* Considerando apenas os 39 campos não *custom field*.
