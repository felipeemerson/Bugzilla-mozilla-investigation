# Contagem* de modificações de campos por valor final de algum campo específico
\* Considerando apenas os 39 campos não *custom field*.

**Observação**: Como a feature product não foi baixada inicialmente com a base, ao baixar tal feature para todos os bug reports da base, o bug report de id 1604167 ficou de fora desta análise por ter passado a ser privado.

## Importações

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline

plt.style.use('ggplot')
pd.set_option('display.float_format', lambda x: '%.2f' % x)


df_bugs_changes = pd.read_json('processed_bugs_counting_changes_final.json')
df_bugs = pd.read_json('bugs_id_product_final.json')
df_bugs.rename(columns={'id': 'bug_id'}, inplace=True)
df_bugs.rename(columns={'product': 'final_product'}, inplace=True)

df_bugs = df_bugs_changes.merge(df_bugs, how='inner', on='bug_id')
df_bugs = df_bugs.drop(columns=['bug_id'])
df_bugs['total_changes'] = df_bugs.sum(axis=1, numeric_only=True)
```

## Agrupamento de contagem de modificações por produto final

**Observação**: Existem 163 produtos no total, foram escolhidos os 10 mais presentes na base

```python
df_bugs.groupby(['final_product'])['total_changes'].describe().reset_index().sort_values('count', ascending=False).head(10).reset_index(drop=True)
```
